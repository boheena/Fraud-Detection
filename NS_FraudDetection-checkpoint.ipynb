## Decision Tree and Random Forest Classification

```python
# apply decision tree to the unbalanced data
feature = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17',
          'V18', 'V19', 'V20', 'V21','V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28']
df['Amount'] = StandardScaler().fit_transform(df['Amount'].values.reshape(-1,1))
X = df[feature]
y = df.Class

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 1)

tree = DecisionTreeClassifier()

tree.fit(X_train,y_train)

y_tree_pred = tree.predict(X_test)
```

```python
print('Tree Score is', tree.score(X_test, y_test))

print('Tree accuracy is', metrics.accuracy_score(y_test, y_tree_pred))

print('Tree precision score is', metrics.precision_score(y_test, y_tree_pred))

print('Tree recall score is ', metrics.recall_score(y_test, y_tree_pred))

print('Tree F1 score is ', metrics.f1_score(y_test, y_tree_pred))

print(metrics.confusion_matrix(y_test, y_tree_pred))
```

```python
# use cross validation 5 folds on Random Forest
random_forest = RandomForestClassifier(n_estimators=70, random_state=0)

random_forest.fit(X_train, y_train)

y_random_forest_pred = random_forest.predict(X_test)

#F1_scores = cross_val_score(random_forest, X, y, cv=5, scoring = 'f1')
```

```python
# print scores for random forest classification
print('Random forest classification Score is', random_forest.score(X_test, y_test))

print('Random forest classification accuracy is', metrics.accuracy_score(y_test, y_random_forest_pred))

print('Random forest classification precision score is', metrics.precision_score(y_test, y_random_forest_pred))

print('Ramdom forest recall score is', metrics.recall_score(y_test, y_random_forest_pred))

print('Random forest f1 score is', metrics.f1_score(y_test, y_random_forest_pred))

print(metrics.confusion_matrix(y_test, y_random_forest_pred))

fpr_rf, tpr_rf, _rf = metrics.roc_curve(y_test, y_random_forest_pred)

plt.plot(fpr_rf, tpr_rf)
plt.title('ROC for Random Forest')
plt.xlabel('Fales Positive Rate')
plt.ylabel('True Positiove Rate')
plt.show()
```

# Using Down-Sampling

```python
# down-sampling the non-fraud data, but would casue a huge information loss

df_non_fraud = df[df.Class == 0]
df_fraud = df[df.Class == 1]

# df_non_fraud['Class'].value_counts()

df_non_fraud_downsample = resample(df_non_fraud, replace=False, n_samples=492, random_state=42)

# df_non_fraud_downsample['Class'].value_counts()

# Combine non-fraud and fraud data together 
df_downsample = pd.concat([df_non_fraud_downsample, df_fraud])

print(df_downsample['Class'].value_counts())
```

## Logistic Regression

```python
#normolize amount

X_downsample = df_downsample[feature]
y_downsample = df_downsample.Class
X_ds_train, X_ds_test, y_ds_train, y_ds_test = train_test_split(X_downsample, y_downsample, test_size = 0.3, random_state = 42)
```

```python
penalty = ['l1', 'l2']
C = np.logspace(0, 4, 10)
hyperparameters = dict(C=C, penalty=penalty)

lr = LogisticRegression()
lr_best = GridSearchCV(lr, hyperparameters, cv=5, verbose=0)
best_model = lr_best.fit(X_ds_train, y_ds_train)
y_lr_pred = best_model.predict(X_ds_test)

print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
```

```python
print('logistic Regreeion Score is', best_model.score(X_ds_test, y_ds_test))

print('logistic regreesion accuracy is', metrics.accuracy_score(y_ds_test, y_lr_pred))

print('logistic regreesion precision score is', metrics.precision_score(y_ds_test, y_lr_pred))

print('logistic regreesion recall score is ', metrics.recall_score(y_ds_test, y_lr_pred))

print('logistic regreesion F1 score is ', metrics.f1_score(y_ds_test, y_lr_pred))

print(metrics.confusion_matrix(y_ds_test, y_lr_pred))

fpr_lr, tpr_lr, threadhold_lr = metrics.roc_curve(y_ds_test, y_lr_pred)

plt.plot(fpr_lr, tpr_lr)
plt.title('ROC for Logistic Regression')
plt.xlabel('Fales Positive Rate')
plt.ylabel('True Positiove Rate')
plt.show()
```
